```python
from ydata_profiling import ProfileReport
```

```python
import pandas as pd
datos = pd.read_csv("/home/laptop/Proyectos Python/LimpiezaDatos/LimpiezaDatos/datos/house_prices.csv", sep=";")
```

```python
#profile = ProfileReport(datos, title='Datos de churn.csv')
#profile.to_notebook_iframe()# Export the report as html and display
from IPython.display import HTML, display
report = ProfileReport(
    datos,
    title="House Prices",
    infer_dtypes=True,
    interactions=None,
    missing_diagrams=None,
    correlations={
        "auto": {"calculate": False},
        "pearson": {"calculate": True},
        "spearman": {"calculate": True},
    },
)
report_html = report.to_html()
display(HTML(report_html))
```
